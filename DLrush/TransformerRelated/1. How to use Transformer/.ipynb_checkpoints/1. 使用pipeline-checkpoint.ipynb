在机器学习中，尤其是在分类任务上，混淆矩阵（Confusion Matrix）是一个非常重要的性能评估工具。情感分析模型，作为一个分类模型的例子（通常是二分类，将情感分为正面或负面），同样可以通过混淆矩阵来评估其性能。

混淆矩阵提供了以下关键信息：

真正类（True Positives, TP）：模型正确预测为正类的实例数。
假正类（False Positives, FP）：模型错误预测为正类的实例数。
真负类（True Negatives, TN）：模型正确预测为负类的实例数。
假负类（False Negatives, FN）：模型错误预测为负类的实例数。
通过这四个基本的度量，可以计算出更多的性能指标，如准确度（Accuracy）、精确率（Precision）、召回率（Recall）和F1分数。这些指标能够提供比简单的准确度更全面的模型性能评估：

准确度告诉我们模型正确预测的实例占总实例的比例。
精确率是模型预测为正类的实例中实际为正类的比例。
召回率是模型正确预测为正类的实例占所有实际正类实例的比例。
F1分数是精确率和召回率的调和平均，用于在两者之间提供一个平衡的性能度量。
在情感分析中，计算混淆矩阵尤为重要，因为它可以揭示模型在预测不同类别时的偏差。例如，模型可能倾向于将多数实例预测为正面情感，而忽视了负面情感的准确预测。这种情况下，即使准确度看起来很高，模型的实际性能也可能并不理想，因为它在负面情感上的假负类（FN）可能很高。混淆矩阵能够帮助我们识别这样的问题，并进行更细致的性能分析。

````python
# plot confusion matrix
def plot_cm(cm):
  classes = ["negative","positive"]
  df_cm = pd.DataFrame(cm,index = classes,columns = classes)
  ax = sn.heatmap(df_cm,annot = True,fmt = 'g')
  ax.set_xlabel("Predicted")
  ax.set_ylabel("Target")

plot_cm(cm)
````

````python
def wrap(x):
  return textwrap.fill(x,replace_whitespace = False,fix_sentence_endings = True)
````

x：这是wrap函数的输入参数，预期是一个字符串，即你希望进行包装（换行处理）的文本。

textwrap.fill(text, width=70, **kwargs)：这是textwrap模块中的一个函数，用于将文本text包装到指定宽度width。如果不指定width参数，默认宽度是70。这个函数返回一个新的字符串，原字符串text中的文本被重新组织，以使每一行都不超过指定的宽度，并且尽可能均匀地分配文本到每一行。

replace_whitespace = False：这是textwrap.fill函数的一个关键字参数。当设置为False时，表示在包装文本时不替换空白字符（如空格、制表符等）。默认情况下，textwrap.fill会将文本中的空白字符替换为一个空格，并且会合并多个连续的空白字符。将此参数设置为False意味着保留原文中的空白字符分布。

fix_sentence_endings = True：这也是textwrap.fill函数的一个关键字参数。当设置为True时，textwrap.fill会尝试检测句子的结束（如通过句号.、问号?、感叹号!等标识），并确保句子的结束不会因为换行而被错误地分割到两行。这有助于保持包装后文本的可读性，特别是对于包含完整句子的文本。

总的来说，wrap(x)函数的作用是对输入的字符串x进行包装，使其在不替换空白字符的情况下，且尽量不在句子结束处分行，返回一个新的、格式化后的字符串。这个函数特别适合处理需要保持原有空白字符和句子完整性的文本格式化任务。

````python
# 写一个函数自动替换词语
# 一种方法是将所有得分能够大于5的替换都替换掉，然后重新传入

def refine_text(text, model):
    # 分割文本为单词列表
    words = text.split()
    # 逐个单词尝试替换
    for i in range(len(words)):
        # 创建 mask 文本
        masked_text = " ".join(words[:i] + [model.tokenizer.mask_token] + words[i+1:])
        # 获取模型预测
        predictions = model(masked_text)
        # 如果第一个选择的得分大于0.05，则采纳
        if predictions[0]['score'] > 0.05:
            # 更新文本
            words[i] = predictions[0]['token_str'].replace(' ', '')  # 如果得分大于0.05，则将words列表中当前位置i的单词替换为预测的单词（通过键'token_str'访问），并使用replace(' ', '')方法移除任何可能的空格。
    # 返回更新后的文本
    return " ".join(words)


refined_text = refine_text(doc,mlm)
print(refined_text)
````

# NER with transformers:A simple Application

````python
from transformers import pipeline
ner = pipeline("ner",aggregation_strategy='simple')
#ner = pipeline("ner",aggregation_strategy='simple',device=0)
````

aggregation_strategy='simple' 指的是一种简单的聚合策略，用于处理同一个实体可能由多个子标记（sub-token）组成的情况。由于一些实体可能会被分词器拆分成多个子标记，简单聚合策略会将同一实体的所有子标记合并为一个单一的实体，并为整个实体指定一个标签。

例如，如果模型识别出 "New" 和 "York" 两个相邻的子标记都被标记为地点（LOC），那么使用 simple 聚合策略，管道会将这两个子标记合并为一个实体 "New York" 并标记为 LOC。

聚合策略通常有以下几种：

none：不进行任何聚合，返回模型原始输出的所有实体标记。

simple：将具有相同实体标签的相邻子标记合并为单个实体。

first：仅保留第一个子标记的标签，并忽略同一实体内的其他子标记。

average：计算同一实体内所有子标记的分数平均值，并基于平均分数确定实体的标签。

max：选择同一实体内得分最高的子标记的标签作为整个实体的标签。

简便性：使用 pickle 模块可以轻松地保存复杂的数据结构，如自定义类、Python 中的数据类型（列表、集合、字典等）和对象实例。

Python专用：.pkl 文件被设计为特定于 Python，这意味着它可以精确地保存和恢复 Python 对象和它们的状态。

快速：pickle 模块的序列化和反序列化过程通常比直接使用文本文件快得多。

````python
from nltk.tokenize.treebank import TreebankWordDetokenizer
detokenizer = TreebankWordDetokenizer()

````


在自然语言处理（NLP）中，nltk.tokenize.treebank.TreebankWordDetokenizer 是由 Natural Language Toolkit (NLTK) 提供的一个类，它用于将已经分词（tokenize）的文本序列重新合并（detokenize）为一段连贯的文本字符串。这个过程通常被称为反标记化（detokenization）。

功能和目的
当文本被分词处理后，我们会得到一个单词列表。分词是将文本分解为更容易分析和处理的单元（例如单词、标点符号等）。在很多NLP任务中，如机器翻译、文本生成等，我们在处理完文本后，经常需要将这些单独的单词重新组合成人类可读的文本。这时候就需要用到反标记化。

TreebankWordDetokenizer 会考虑到标点符号和单词之间的空格，将分词后的列表转换成格式正确的句子。它尝试恢复原始句子的空白字符和标点符号，使其看起来像是自然书写的文本。

例如：

```python
tokens = ["This", "is", "an", "example", "of", "text", "detokenization", "."]
detokenizer = TreebankWordDetokenizer()
detokenized_text = detokenizer.detokenize(tokens)
print(detokenized_text)  # This is an example of text detokenization.
```
在这个例子中，detokenize 方法接收一个单词列表 tokens，然后返回一个字符串 detokenized_text，其中单词之间的空格恢复到适当的位置，标点符号也正确地附加到单词后面，使整个句子看起来自然。




在某些情况下，直接使用 join 方法将分词后的列表重新组合成字符串是可行的，特别是当文本处理的需求比较简单，不涉及复杂的标点符号处理或特殊格式要求时。例如：

```python

tokens = ["This", "is", "an", "example", "."]
text = " ".join(tokens)
print(text)  # 输出: This is an example .
```
然而，这种方法有其局限性，主要原因包括：

标点符号处理：直接使用 join 方法时，每个词之间都会被添加一个空格，包括那些应该紧贴前一个词的标点符号（如句号、逗号等）。这会导致文本格式不符合自然语言的书写习惯。

空格规则：在自然语言中，不同语境下单词之间的空格规则可能有所不同。例如，英文中通常在逗号之前不加空格，在逗号之后加空格。直接使用 join 方法无法灵活处理这些情况。

复杂文本格式：对于含有特殊格式要求的文本（如引号内的文本、带括号的表达式等），直接使用 join 方法可能无法正确恢复原文本的格式。

````python
# 考虑到NER模型只会输出你给定的词性，这个函数会填补status为"O"的token，以方便我们计算模型准确度
def compute_prediction(tokens, input_,ner_result):
  # map hugging face ner result to list of tags for later performance assessment
  # tokens：原始句子经过分词处理后得到的token列表。
  # input_：反标记化（detokenized）后的字符串，即原始句子。
  # ner_result：NER模型的输出结果，是一个包含识别出来的实体信息的字典列表，每个字典通常包含实体的类型、起始位置和结束位置。

  # predicted_tags：用于存储每个token对应的实体标签。
  # state：用于跟踪当前实体的状态，'O'表示非实体，'B'表示实体开始，'I'表示实体内部。
  # current_index：用于跟踪input_字符串中当前处理到的位置。
  predicted_tags = []

  state = 'O' # keep track of state, so if O-->B,if B --> I, if I-->I
  current_index = 0

  # 对于tokens列表中的每个token，在input_字符串中找到它的位置index。
  # 更新current_index以指向input_中对应token的起始位置。
  for token in tokens:
    # 从当前索引位置开始查找token
    index = input_.find(token)
    assert(index >= 0)
    current_index = index # 更新current_index到找到的token的起始位置

    # print(token,current_index)

    # 初始化tag为'O'，表示当前token默认不属于任何实体。
    tag = 'O'
    # 遍历ner_result中的每个实体，检查current_index是否在实体的起始和结束位置范围内。
    for entity in ner_result:
      # 如果在范围内，根据state更新tag为'B-实体类型'或'I-实体类型'，并更新state。
      if current_index >= entity['start'] and current_index < entity['end']:
        # then this token belongs to an entity
        if state=='O':
          state = 'B'
        else:
          state = 'I'

        tag = f"{state}-{entity['entity_group']}"
        break




    if tag == 'O':
      state='O'


    # 将tag添加到predicted_tags列表。
    predicted_tags.append(tag)

  # 断言predicted_tags和tokens的长度相等，确保每个token都有对应的标签。
  assert(len(predicted_tags) == len(tokens))

  return predicted_tags



````

````python
# 以上代表模型work了，接下来我们应该对整个数据集使用
# get detokenized inputs to pass into ner model
detok_inputs = []
for tokens in inputs:
  text = detokenizer.detokenize(tokens) # detokenize each sentenance
  detok_inputs.append(text)
````

````python
ner_results = ner(detok_inputs)# pass our detokenized input into pipeline
````

````python
# 使用我们之前用过的函数补全整个句子的预测
predictions = []
for tokens,text,ner_result in zip(inputs,detok_inputs,ner_results):
  pred = compute_prediction(tokens,text,ner_result)
  predictions.append(pred)
````

````python
# https//stackoverflow.com/questions/11264684/flatten-list-of-lists
# 展平这个list of list，即为提取每个list的每个词按顺序组成list
def flatten(list_of_lists):
  flattened = [val for sublist in list_of_lists for val in sublist]
  return flattened
````

````python
# flatten targets and predictions
flag_predictions = flatten(predictions)
flat_targets = flatten(targets)
````

````python
accuracy_score(flat_targets,flag_predictions)
````

````python
f1_score(flat_targets,flag_predictions,average='macro')
````

左侧是英文文本，右侧是Spanish，可以发现，对于单个输入，存在潜在的多个可能输出

````python
# load file and store dataset in a suitable structure
eng2spa = {} # dictionary

for line in open('spa-eng/spa.txt'): #对于每一行
    # 去除尾部空格
    line = line.rstrip() 
    # 按制表符分离
    eng,spa = line.split("\t")

    # multi-unordered map
    if eng not in eng2spa:
        eng2spa[eng] = []
    eng2spa[eng].append(spa)
````

````python
# compute blue score
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
````

````python
# 翻译任务特有的tokenizer
from nltk.tokenize import RegexpTokenizer
````

````python
tokenizer = RegexpTokenizer(r'\w+')
````

````python
tokenizer.tokenize('¿Quién eres?'.lower())
#"你是谁"的西班牙语
````

````python
tokens = tokenizer.tokenize('¿Quién eres?'.lower())
# []是可能的翻译，这里尝试将自己作为自己的翻译，
sentence_bleu([tokens],tokens)
````

获得一个接近于0的BLEU分数（在这个例子中是1.491668146240062e-154）可能是由几个因素导致的。理论上，当参考翻译和候选翻译完全相同时，BLEU分数应该是1（或100%），表示完美匹配。然而，实际上BLEU分数的计算涉及多个层面的匹配，包括单词级别（1-gram）到更长序列的匹配（比如2-gram, 3-gram等），以及一个用于避免过短翻译的惩罚因子。如果在较长的n-gram匹配上得分较低，或者出现了长度惩罚，即使文本看似相似，BLEU分数也可能接近于0。

BLEU分数的计算方法
BLEU分数的计算基于几个关键组成部分：

n-gram精确度：计算候选翻译中的n-gram与任一参考翻译中的n-gram匹配的程度。随着n的增加，匹配越来越难，尤其是在参考翻译数量较少的情况下。

惩罚因子：如果候选翻译的长度小于参考翻译的最佳长度，将应用一个惩罚因子以避免过短的翻译。

平滑函数：对于完全没有匹配的n-gram，传统的BLEU分数计算方法会直接返回0。为了避免这种极端情况，通常会应用平滑函数来给予一些微小的分数，使得总体BLEU分数不会直接为0。

导致低BLEU分数的原因
在你的例子中，尽管参考翻译和候选翻译看似相同，但还是可能因为以下原因导致BLEU分数异常低：

n-gram匹配度低：如果候选翻译缺少足够的高阶n-gram匹配（比如2-gram及以上），即使1-gram匹配度很高，总体BLEU分数也可能很低。
平滑函数：使用的平滑函数可能对于完全匹配的情况不够敏感，导致分数接近于0。

````python
smoother = SmoothingFunction()
sentence_bleu(['hi'], 'hi', smoothing_function=smoother.method4)
````

````python
# 也能适用于integer
sentence_bleu([[1,2,3,4]],[1,2,3,4])
````

````python
# pretokenizer targets
# 将之前的spanish-English字典转换为tokenize字典
eng2spa_tokens = {} #字典

for eng, spa_list in eng2spa.items():
    spa_list_tokens = []
    for text in spa_list:
        tokens = tokenizer.tokenize(text.lower())
        spa_list_tokens.append(tokens)
    eng2spa_tokens[eng] = spa_list_tokens
````

````python
from transformers import pipeline
translator = pipeline("translation", model='Helsinki-NLP/opus-mt-en-es',device=0)
````

````python
translator("I like eggs and ham")
````

````python
# 查看eng2spa有多少个短语
eng_phrases = list(eng2spa.keys())
len(eng_phrases)
````

````python
eng_phrases_subset = eng_phrases[20000:21000]
````

````python
eng_phrases_subset
````

````python
translations = translator(eng_phrases_subset)
````

````python
translations[0]
````

````python
# 测试翻译成果
scores = []
for eng, pred in zip(eng_phrases_subset, translations):
    matches = eng2spa_tokens[eng]

    spa_pred = tokenizer.tokenize(pred['translation_text'].lower())

    score = sentence_bleu(matches,spa_pred)
    scores.append(score)
````

````python
import matplotlib.pyplot as plt
````

````python
plt.hist(scores, bins=50);
````

````python
import numpy as np
np.mean(scores)
````

````python
# 旨在随机选择一个英语短语，并打印出对应的西班牙语翻译
def print_random_translation():
    i = np.random.choice(len(eng_phrases_subset))
    eng =eng_phrases_subset[i]
    print("EN:",eng)

    translation = translations[i]['translation_text']
    print("ES translation:", translation)

    matches = eng2spa[eng]
    print("Matches:", matches)
````

````python
from transformers import pipeline
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
import textwrap

from sklearn.metrics import roc_auc_score, f1_score, confusion_matrix
````

````python
classifier = pipeline("zero-shot-classification",device=0)
````

````python
classifier("This is a great movie", candidate_labels=["positive","negative"])
````

````python
text = "Due to the presence of isoforms of its components, there are 12 "+ \
        "versins of AMPK in mammals, each of which can have difference tissue " + \
        "localizations, and different functions under different conditions."+\
        "AMPK is regulated allosterically and by post-translational " +\
        "modification, which work together."
classifier(text, candidate_labels = ["biology", "math", "geology"])
````

````python
df = pd.read_csv('bbc_text_cls.csv')
````

````python
# set去重，list转化为列表，我们最终得到了一个不重复的label列表
labels=list(set(df['labels']))
labels
````

````python
print(textwrap.fill(df.iloc[1024]['text']))
````

````python
classifier(df.iloc[1024]['text'],candidate_labels=labels)
````

模型预测为Politics，但是其置信值却不高

````python
# 对所有例子检测判定准确性

preds = classifier(df['text'].tolist(), candidate_labels=labels)
````

````python
preds
````

````python
# 选取预测概率最大的标签
predicted_labels = [d['labels'][0] for d in preds]
````

````python
# 将预测结果写回csv
df['predicted_labels']=predicted_labels
````

````python
print("Acc:", np.mean(df['predicted_labels']==df['labels']))
````

````python
# 列出每个案例的每个标签的概率，label2idx = num_news * num_labels
N = len(df)
K = len(labels)
label2idx = {v:k for k,v in enumerate(labels)} #将label转化为0，1，2，3，4并建立hash表

probs = np.zeros((N,K))

for i in range(N):
    d = preds[i]
    for label, score in zip(d['labels'],d['scores']):
        k = label2idx[label]# probability k
        probs[i,k] = score 
````

````python
int_labels = [label2idx[x] for x in df['labels']] #将原始csv的label转化为索引index存在int_labels
````

````python
# 计算混淆矩阵
int_preds = np.argmax(probs,axis=1)
cm = confusion_matrix(int_labels, int_preds, normalize='true')
````

````python
def plot_cm(cm):
    df_cm = pd.DataFrame(cm,index=labels,columns=labels)
    ax = sn.heatmap(df_cm,annot=True,fmt='.2g')
    ax.set_xlabel("Predicted")
    ax.set_ylabel("Target")

plot_cm(cm)
````

````python
f1_score(df['labels'],predicted_labels,average='micro')
````

````python
roc_auc_score(int_labels,probs,multi_class='ovo')
````
